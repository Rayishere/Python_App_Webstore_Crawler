```python
def app_info(site):
    req = urllib2.Request(site)
    req.add_header("User-Agent","Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) "\
                   "Chrome/51.0.2704.79 Safari/537.36 Edge/14.14393")
    #Open website
    website = urllib2.urlopen(req)
    #Read in website
    content = website.read()
    #findout what kind of content-type
    encoding = website.headers['content-type'].split('charset=')[1]
    #convert to unicode
    ucontent = unicode(content, encoding)
    
    
    soup = BeautifulSoup(ucontent, "lxml")
    
    title = soup.find("div", "main-content").find("h1").text[1:]
    dev = soup.find("div", "left-info").find("a").text[1:]
    app_catagory = soup.find("div","left-info").find_all("a")[1].text[1:]
    app_rate = soup.find("div", "details-info").find("img", "document-subtitle content-rating-badge")['alt']
    star = soup.find("div", "tiny-star star-rating-non-editable-container")['aria-label'][4:10]
    
    
    
    #star_people = soup.find("span", "rating-count")['aria-label']
    
    
    '''
    if(star_people == None):
        star_people = convert.toString(0)
        next
    '''
    app_list = [title, dev, app_catagory, app_rate, star, star_people]
    
    for x in app_list:
        print x
    
app_info("https://play.google.com/store/apps/details?id=com.youmusic.magictiles") 
```
