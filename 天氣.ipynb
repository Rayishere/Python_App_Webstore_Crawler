```python
import requests
from bs4 import BeautifulSoup
import lxml

res=requests.get("https://play.google.com/store/apps/category/WEATHER?hl=zh_TW")
soup=BeautifulSoup(res.text, "lxml")

print "天氣類熱門付費下載 ",'\n'
for title in soup.select('.details'):
    link=title.select('.title')[0]
    print "天氣名稱：",title.select('.title')[0].text
    
    res2=requests.get("https://play.google.com"+str(link['href']))
    soup2=BeautifulSoup(res2.text)

    expand=soup2.select('.screenshot')
    for screen in expand:
        print 'https:'+screen['src']

    details=soup2.select('.show-more-content')[0].text
    print "天氣介紹：",details

    scores=soup2.select('.score')[0].text
    print "分數：",scores
    print'\n'
```
